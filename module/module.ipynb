```python
#date time detector

import re

for i in alpha_columns:
    for j in time_columns:
        pattern = (
            r'(^' + re.escape(j) + r'$|' +
            r'^' + re.escape(j) + r'([_\-\s]|$)|' +
            r'([_\-\s]|^)' + re.escape(j) + r'$|' +
            r'.*' + re.escape(j) + r'.*'
            r')'
        )
        if re.search(pattern, i, re.IGNORECASE):
            if i not in date_columns:
                date_columns.append(i)
            break

def detect_and_convert_date(col):
    col = col.astype(str).str.strip()
    if col.str.fullmatch(r"\d{1,2}").all():
        return col.astype(int), None, None
    if col.str.fullmatch(r"\d{2,4}").all():
        return col.astype(int), None, None
    parsed = pd.to_datetime(col, errors="coerce", format="%d-%m-%Y")
    month_col = parsed.dt.month
    year_col = parsed.dt.year
    return parsed.dt.strftime("%d-%m-%Y"), month_col,year_col

for r in date_columns:
    df[r], month,year = detect_and_convert_date(df[r])
    if month is not None:
        df[r + "_month"] = month
        date_columns.append(r + "_month")
        
    if year is not None:
        df[r + "_year"] = year
        date_columns.append(r + "_year")

    if r in alpha_columns:
        alpha_columns.remove(r)
```
